```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[2,3,4,5],'r*-') # (x, y, plot property)
plt.axis([0,5,0,7]) # Setting the max and minimum values
plt.show()
```

```python
# Bar chart
plt.bar(["Vijayaraghavan","Yashwanth","Kutty"],[30,40,90],width=0.4)

plt.show()
```

```python
# Pie Chart
plt.pie([50,60,10],labels=["Vj","VR","VjR"],explode=[0,0,0.1])
plt.show()
```

```python
#Scatter Plot
plt.scatter([1,2,3,4,5,6,7,8,3,5,7],[3,2,9,4,6,1,8,4,2,7,5],s=50,c='red',alpha=0.9)
# x,y , s- size, c-color, alpha-opacity of the particles
```
